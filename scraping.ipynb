```python
os.chdir('F:/My_Disk/python/Projects/NLP/output_data/')
websites_having_different_structure = []

for index, row in input_url_df.iterrows():
    url = row['URL']
    url_id = row['URL_ID']

    page = requests.get(url)
    # if page.status_code == 200:
    soup = BeautifulSoup(page.content, "html.parser")
    try: 
        res = soup.find("h1", class_="entry-title").text.strip()
        res2 = soup.find("div", class_= "td-post-content tagdiv-type").text.strip()
        res3 = res + ". "+ res2

        with open(f'{url_id}.txt','w', encoding='utf-8') as text_file:
            text_file.write(res3)
    except:
        websites_having_different_structure.append(url_id)
        
    print(f'{index} url scraped')
```

```python
websites_having_different_structure # list of those urls that have different website structure and will be scrapped differently
```

```python
df = input_url_df[input_url_df['URL_ID'].isin(websites_having_different_structure)] # make a df of those url and their id's
print('Number of additional urls to be scraped: ',len(df))
for index, row in df.iterrows():
    url = row['URL']
    id = row['URL_ID']
    print(f'{index}: {id}: {url}')
```
